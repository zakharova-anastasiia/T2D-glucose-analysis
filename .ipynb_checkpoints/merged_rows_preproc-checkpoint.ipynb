```python
import pandas as pd
import numpy as np
from datetime import timedelta
import matplotlib.pyplot as plt
```

```python
df_clear = pd.read_csv('V3_redundant_columns_removed.csv')
df_clear.shape[0] - df_clear.drop_duplicates().shape[0]
```

```python
# sort by col Date
df_p = df_clear.sort_values('Date')
```

```python
# first event date for every patient
pat_start_date = {pat: min(df_p[df_p.Patient_ID == pat].Date) for pat in df_p.Patient_ID.drop_duplicates()}
```

```python
# Dates in terms of days from first event
df_p['Date_transformed'] = pd.Series([pd.to_datetime(df_p.Date[i]) - 
                                      pd.to_datetime(pat_start_date[df_p.Patient_ID[i]])
                                      for i in range(df_p.shape[0])])
```

```python
# dist with keys of patients who has Glu (vein) test (their own dataframes as values)
have_glu_test = {}
for pat in df_p.Patient_ID.drop_duplicates():
    res = df_p[(df_p['Glu (vein)'] != 0) & (df_p.Patient_ID == pat)]
    if res.shape[0] != 0:
        have_glu_test[pat] = res
```

```python
# how many patients had Glu test at least once, all patients 
(len(have_glu_test.keys()), len(df_p.Patient_ID.drop_duplicates()))
```

```python
# for every patient who made test (Glu) dates when they did it 
dates_when_test_was_made = {pat: df_p[(df_p.Patient_ID == pat) & (df_p['Glu (vein)'] != 0)].Date
                           for pat in have_glu_test.keys()}
```

```python
# still 207 duplicates as in beginning (i think that these duplicates are mistakes)
df_p.shape[0] - df_p.drop_duplicates().shape[0]
```

```python
# how many times a day patiens had their Glu tests
freq = []
for pat in have_glu_test.keys():
    freq.append(have_glu_test[pat].Date.value_counts()[0])
```

```python
# hist of how many times a day patiens had their Glu tests
plt.hist(freq, bins = len(set(freq)))
plt.title('Times per day Glucose test was made')
plt.xlabel('times')
```
